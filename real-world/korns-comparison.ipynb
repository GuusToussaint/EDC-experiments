```python
import os
import pickle

results_files = os.listdir("comparison_results")

results = []
for result_file in results_files:
    if "beam" not in result_file:
        continue
    dataset_type, dataset, search_strategy, optimiser, random_seed = result_file.split(
        "."
    )[0].split("-")
    with open(f"comparison_results/{result_file}", "rb") as f:
        res = pickle.load(f)
    for elapsed_time, auc_score, accuracy_score in zip(res["elapsed_time"], res["auc_score"], res["accuracy_score"]):
        results.append(
            [
                dataset,
                "EDC",  # "classifier
                search_strategy,
                optimiser,
                elapsed_time,
                auc_score,
                accuracy_score,
                random_seed,
            ]
        )
```

```python
import numpy as np

T1_results = list(filter(lambda x: x[0] == "T1", results))
T2_results = list(filter(lambda x: x[0] == "T2", results))
T3_results = list(filter(lambda x: x[0] == "T3", results))

accuracy_score = np.mean([x[6] for x in T1_results])
print(f"T1 average CEP: {1-accuracy_score}")

accuracy_score = np.mean([x[6] for x in T2_results])
print(f"T2 average CEP: {1-accuracy_score}")

accuracy_score = np.mean([x[6] for x in T3_results])
print(f"T3 average CEP: {1-accuracy_score}")
```
