```python
# Imports

import os
import pandas as pd
import matplotlib.pyplot as plt
import pickle
```

```python
# Load the results

num_iterations = 100

base_path = "/data1/toussaintg1/edc-2-optimisers/"
result_files = os.listdir(base_path)
results = []
for result_file in result_files:
    optimiser, random_seed = result_file.split(
        "."
    )[0].split("-")
    with open(f"{base_path}{result_file}", "rb") as f:
        res = pickle.load(f)
    for accuracy_score, auc_score in zip(res["accuracy_score"], res["auc_score"]):
        results.append(
            [
                optimiser,
                random_seed,
                auc_score,
                accuracy_score
            ]
        )

df = pd.DataFrame(
    columns=[
        "optimiser",
        "random_seed",
        "auc_score",
        "accuracy_score"
    ],
    data=results,
)

name_dict = {
    "hill_climber": "Hill Climber",
    "hill_climber_optimised": "Hill Climber",
    "hill_climber_gradient_decent": "Gradient Descent + HC",
    "hill_climber_gradient_decent_v2": "Gradient Descent + HC",
    "hill_climber_gradient_decent_v3": "Gradient Descent + HC",
    "random_sample": "Random Sample",
    "gradient_decent": "Gradient Descent + RS",
    "random_search": "Random Search",
    "random_search_gradient_decent": "Random Search Gradient Descent",
    "nelder_mead": "Nelder-Mead",
    "powell": "Powell",
}

grouped = df.groupby(["optimiser"])
```

```python
# Print the results
sorted_grouped = sorted(grouped, key=lambda group: group[1]["auc_score"].mean(), reverse=True)

for name, group in sorted_grouped:
    mean_auc = group["auc_score"].mean()
    std_auc = group["auc_score"].std()
    name = name[0]
    if name in name_dict:
        name = name_dict[name]
    print(f"{name}&{mean_auc:.4f} ($\\pm{std_auc:.2f}$)\\\\")
```
